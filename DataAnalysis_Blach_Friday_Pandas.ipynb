**Handling Missing Values **

Pandas provides isna() and notna() functions to detect "NA" Values

detect NA values in the dataframe:  df.isna().sum()

detect NA values in particular column in the dataframe: pd.isna(df["col_name"]) . df["col_name"].notna()

isnull() is just an alias of the isna()

```python
df.isnull().sum()
# as you can see in the Product_Category_2  and Product_Category_2 have null values
```

```python
df.Product_Category_2
# Pandas always recognize missing values as NAN
```

Handle mising values using dropping and imputing both options.

1. Dropping

2. Imputing (Mean, Median, Mode,

```python
# Dropping Product_Category2 using axis = 1
"""axis=1: This specifies that you are dropping a column (along the horizontal axis).
inplace=False: This means that the operation will not modify the original DataFrame """

df_Temp = df.drop("Product_Category_2", axis = 1, inplace = False)
df_Temp
```

```python
df_Temp.info()
```

```python
df_Temp.isnull().sum()
```

```python
# Get index for all rows with Product_Category_3 missing

df_Temp[df_Temp.Product_Category_3.isnull()].index
```

```python
# delete all missing values from Product_Category_3,  # We can use dropna() function to remove all Nan (null) value
df_Temp = df_Temp.drop(df_Temp[df_Temp.Product_Category_3.isnull()].index, axis = 0, inplace = False)
```
